```python
boat_side = 'Right'
missionaries_on_right = 3
cannibals_on_right = 3
missionaries_on_left = 0
cannibals_on_left = 0

print('Right Side')
print('M = '+str(missionaries_on_left) +' C = ' + str(cannibals_on_left) + '|-----B|' + 'M = ' + str(missionaries_on_right) + ' C = ' +str(cannibals_on_right))

while True:
   
#boat on right side
  if boat_side == 'Right':   # using if-else loop check the number of missionaries and cannibles when boat is on the right side.
    
    missionaries = int(input('ENTER THE NUMBER OF MISSIONARIES ON BOAT: '))    #take input from the user 
    cannibals = int(input('ENTER THE NUMBER OF CANNIBALS BOAT: '))   #take input from the user
    
    if (missionaries + cannibals)<0 or (missionaries + cannibals)>2:    # input taken should be atleast 1 or 2,else print invalid move
      print('Invalid Move 1')
      continue
    else:
      if (missionaries_on_right < missionaries) or (cannibals_on_right < cannibals):
        print('Invalid Move 2')
        continue
      else:
        cannibals_on_right -= cannibals
        missionaries_on_right -= missionaries
        cannibals_on_left += cannibals
        missionaries_on_left += missionaries

        print('M = '+str(missionaries_on_left) +' C = ' + str(cannibals_on_left) + '|B-----|' + 'M = ' + str(missionaries_on_right) + ' C = ' +str(cannibals_on_right))
        boat_side = 'Left'

#boat on left side
  elif boat_side == 'Left':

    missionaries = int(input('ENTER THE NUMBER OF MISSIONARIES ON BOAT: '))    #Take input from the user
    cannibals = int(input('ENTER THE NUMBER OF CANNIBALS BOAT: '))  #Take input from the user
    
    if (missionaries + cannibals) < 0 or (missionaries + cannibals) > 2 :   # input taken should be atleast 1 or 2,else print invalid move
      print('Invalid Move 1')
      continue
    else:
      if (missionaries_on_left < missionaries) or (cannibals_on_left < cannibals):
        print('Invalid Move 2')
        continue
      else:
        cannibals_on_left -= cannibals
        missionaries_on_left -= missionaries
        cannibals_on_right += cannibals
        missionaries_on_right += missionaries

        print('M = '+str(missionaries_on_left) +' C = ' + str(cannibals_on_left) + '|-----B|' + 'M = ' + str(missionaries_on_right) + ' C = ' +str(cannibals_on_right))
        boat_side = 'Right'

  #Game result
  if ((missionaries_on_right < cannibals_on_right) and (missionaries_on_right > 0)) or ((missionaries_on_left < cannibals_on_left) and (missionaries_on_left > 0)):   #if no.of missionaries on right is less than no. of cannibals on right and no.of missionaries on right is greater than 0 you lose the game.
    print('YOU LOSE')    # or if no.of missionaries on left is less than no. of cannibals on left and no.of missionaries on left is greater than 0 you lose the game.
    break        #if above condition is satisfied you lose the game,thus by using break function game stops on this point and need to start the new game.
    
  elif (missionaries_on_left and cannibals_on_left) == 3:    # if no.of missionaries on left and no. of cannibals on left==3 you win the game
    print('YOU WIN')  
    break  #by using break function game stops here.

print('GAME OVER')  #if any of the above condition is satisfied "GAME OVER" wll print
```
